# Predição do Tamanho de Pelotas

**Processo Industrial de Pelotização**

O processo de pelotização foi desenvolvido para o aproveitamento de minérios concentrados, ultrafinos, impróprios para o uso direto nos fornos siderúrgicos de produção de ferro primário, tais como: alto-forno, reator de redução direta, corex, etc.

A pelotização consiste na aglomeração desses finos, transformando-os em esferas de **diâmetro médio da ordem de 12mm**, com propriedades químicas, físicas e metalúrgicas adequadas para uso, são mais fáceis de manusear, transportar em várias aplicações.

**Processo Industrial de Pelotização**

- **Preparação das Matérias-Primas:** A etapa inicial inclui a britagem e moagem das matérias-primas para alcançar um tamanho de partícula fino adequado para a pelotização. Para o minério de ferro, isso geralmente significa garantir que cerca de 85% das partículas tenham menos de 44 micrômetros.

- **Mistura com Agentes Ligantes:** Os materiais moídos são misturados com água e um agente ligante (como bentonita para minério de ferro) para aumentar a coesão entre as partículas. Essa mistura é crucial para formar pelotas que possam suportar manuseio e processamento.

- **Formação das Pelotas:** O material misturado é então alimentado em um pelotizador, que pode ser um disco ou um tambor. Nesses dispositivos, a mistura é agitada, fazendo com que as finas se unam e formem pequenos aglomerados conhecidos como pelotas-semente. À medida que essas pelotas-semente se movem umas contra as outras, elas crescem ao absorver finos adicionais até atingirem o tamanho desejado.

- **Secagem e Processamento Térmico:** Após a formação das pelotas, elas passam por um processo de secagem para reduzir o teor de umidade e por tratamento térmico (sinterização) para melhorar suas propriedades mecânicas. Isso geralmente envolve a queima das pelotas em altas temperaturas (cerca de 1250-1300°C) para endurecê-las.

- **Resfriamento e Triagem:** Finalmente, as pelotas são resfriadas e triadas para separar aquelas que atendem às especificações de tamanho daquelas que não atendem. Pelotas oversized ou undersized são recicladas de volta ao processo.

**Principais Aplicações no Contexto Industrial**

- **Produção de Ferro e Aço:** Pelotas de minério de ferro são usadas como matéria-prima em altos-fornos devido às suas propriedades físicas e metalúrgicas melhoradas em comparação com o minério bruto.

- **Energia Biomassa:** Pelotas de madeira produzidas a partir de serragem e aparas de madeira servem como fontes renováveis de energia para aquecimento e geração de eletricidade.

- **Fertilizantes:** Fertilizantes peletizados garantem uma distribuição uniforme em aplicações agrícolas, melhorando a entrega de nutrientes às culturas.

- **Indústria Química:** Vários produtos químicos e minerais (como gesso e calcário) são pelotizados para melhor manuseio e aplicação em processos industriais.

- **Melhoradores do Solo:** Materiais orgânicos peletizados melhoram a qualidade do solo ao fornecer nutrientes de maneira controlada.

Como já constatado na análise exploratória preliminar com o YData Profiling, existem muitos outliers na base. Por exemplo, se o sensor **capturou o valor 0 na variável "Taxa_Alimentação_Misturador"** durante o processo produtivo, isso é um indicativo claro de que houve algum problema na medição. Nesse caso específico, **optarei por excluir da base de dados os valores discrepantes**, pois não fazem sentido no contexto do problema de negócio em questão. Essa abordagem visa garantir a qualidade dos dados e a precisão das análises subsequentes.


```python
# Função para remover linhas que contenham valores iguais a 0 e próximos de 0

def remover_zero_e_proximos(df, threshold=0.05):

    # Fazendo uma cópia do DataFrame original para não alterá-lo

    df_tratado = df.copy()

    # Selecionando apenas as colunas numéricas

    df_numerico = df_tratado.select_dtypes(include=np.number)

    # Criando uma máscara para identificar linhas com valores iguais a 0 ou próximos de 0

    mask = (df_numerico <= threshold) & (df_numerico >= -threshold)

    # Removendo as linhas que contêm valores iguais a 0 ou próximos de 0

    df_tratado = df_tratado[~mask.any(axis=1)]

    return df_tratado

# Exemplo de uso da função

df_tratado = remover_zero_e_proximos(df)
```

Neste bloco de código, eu criei a função `remover_zero_e_proximos`, que tem como objetivo remover linhas de um DataFrame que contêm valores iguais a 0 ou próximos de 0, utilizando um parâmetro ``threshold` para definir a proximidade.

Primeiro, faço uma cópia do DataFrame original para garantir que ele não seja alterado durante o processo. Em seguida, seleciono apenas as colunas numéricas do DataFrame copiado.

Depois, crio uma máscara que identifica as linhas que têm valores dentro do intervalo definido pelo `threshold`, ou seja, valores que são menores ou iguais a 0.05 e maiores ou iguais a -0.05.

Com essa máscara, removo as linhas que contêm esses valores, mantendo apenas aquelas que não atendem à condição. Por fim, a função retorna o DataFrame tratado, que agora não contém linhas com valores iguais a 0 ou próximos de 0.

No exemplo de uso, aplico a função ao DataFrame `df` e armazeno o resultado em `df_tratado`.

**Por que é importante a exclusão de valores 0 e próximos de 0 ?**

A exclusão de valores iguais a 0 e próximos de 0 é fundamental para garantir a precisão das métricas, especialmente o MAPE (Mean Absolute Percentage Error). Quando os valores reais são zero, a fórmula do MAPE resulta em uma divisão por zero, o que pode gerar erros ou resultados indefinidos.

Além disso, valores muito próximos de zero podem distorcer o cálculo, fazendo com que pequenos erros nas previsões resultem em altos percentuais de erro, dando uma impressão enganosa sobre o desempenho do modelo. Isso pode levar a uma interpretação errada da eficácia do modelo, já que o MAPE pode parecer inflacionado.

Ao remover esses valores, consigo focar em dados mais representativos, garantindo que as métricas reflitam com precisão a capacidade do modelo de prever em situações normais. Assim, a exclusão é crucial para uma avaliação mais justa e precisa do desempenho do modelo.

```python
# Dataframe para amazenar as métricas

df_results = pd.DataFrame(columns=['R2', 'MAE', 'RMSE', 'RMSLE', 'Max Error', 'MAPE'])

# Lista de métricas a serem calculadas

metrics = [
                'r2',
                'neg_mean_absolute_error',
                'neg_root_mean_squared_error',
                'neg_root_mean_squared_log_error',
                'max_error',
                'neg_mean_absolute_percentage_error',
]

# Validação cruzada para dividir os dados

kfold = KFold(n_splits=5, shuffle=True, random_state=random_state)

# Iteração sobre modelos

for t, models in all_models.items():
    if t == 'linear_models':
        steps = steps_linear_models
    elif t == 'tree_models':
        steps = steps_em_comum

    for model in models:
        model_name = model[0]
        print(f'Treinando {model_name} ...', end= ' ')

        m_pipe = Pipeline(steps=steps + [model])

        cv_results = cross_validate(estimator=m_pipe, X=X_train, y=y_train, scoring=metrics, cv=kfold, n_jobs=-1)
        cv_results_df = pd.DataFrame(cv_results)

        # Armazenamento dos resultados

        df_results.loc[model_name] = [
            cv_results_df['test_r2'].mean(),
            np.abs(cv_results_df['test_neg_mean_absolute_error'].mean()),
            np.abs(cv_results_df['test_neg_root_mean_squared_error'].mean()),
            np.abs(cv_results_df['test_neg_root_mean_squared_log_error'].mean()),
            np.abs(cv_results_df['test_max_error'].mean()),
            np.abs(cv_results_df['test_neg_mean_absolute_percentage_error'].mean()),
        ]
        print('OK')
```

Neste código, estou criando um DataFrame chamado `df_results` para armazenar as métricas de desempenho dos modelos de regressão. As colunas do DataFrame incluem várias métricas, como R², MAE, RMSE, RMSLE, Erro Máximo e MAPE.

Em seguida, defino uma lista chamada `metrics`, que contém os nomes das métricas que quero calcular durante a validação cruzada. Utilizo o `KFold` para dividir os dados em 5 partes, garantindo que os dados sejam embaralhados a cada execução, o que é controlado pelo `random_state`.

O loop principal itera sobre os grupos de modelos definidos no dicionário `all_models`. Dependendo do tipo de modelo (linear ou árvore), seleciono os passos apropriados para o pipeline. Para cada modelo, crio um pipeline que combina os passos de pré-processamento com o modelo em si.

Em seguida, utilizo a função `cross_validate` para avaliar o modelo, passando o pipeline, os dados de treinamento e as métricas definidas. Os resultados da validação cruzada são armazenados em um DataFrame temporário, `cv_results_df`.

Por fim, extraio as médias das métricas de desempenho e as insiro no DataFrame `df_results`, usando o nome do modelo como índice. O código imprime uma mensagem de confirmação após o treinamento de cada modelo, indicando que o processo foi concluído com sucesso.

```python
# Função para calcular métricas de desempenho

def adjusted_r2(y_esperado, y_previsto, X_treino):
    R2 = r2_score(y_esperado, y_previsto)
    N = len(y_esperado)
    p = X_treino.shape[1]
    r2_ajustado = 1 - ((1 - R2) * (N - 1)) / (N - p - 1)
    return r2_ajustado

# Função para gerar métricas de regressão

def gerar_metricas_regressao(y_test, y_pred_test, y_train, y_pred_train, X_train):

    # Calculando as métricas

    metrics_reg = {
        'Métrica': ['R² Ajustado', 'MAE', 'MSE', 'RMSE', 'RMSLE', 'Max Error', 'MAPE'],
        'Treino': [
            adjusted_r2(y_train, y_pred_train, X_train),
            mean_absolute_error(y_train, y_pred_train),
            mean_squared_error(y_train, y_pred_train),
            mean_squared_error(y_train, y_pred_train, squared=False),
            mean_squared_log_error(y_train, y_pred_train, squared=False),
            max_error(y_train, y_pred_train),
            mean_absolute_percentage_error(y_train, y_pred_train)
        ],
        'Teste': [
            adjusted_r2(y_test, y_pred_test, X_train),
            mean_absolute_error(y_test, y_pred_test),
            mean_squared_error(y_test, y_pred_test),
            mean_squared_error(y_test, y_pred_test, squared=False),
            mean_squared_log_error(y_test, y_pred_test, squared=False),
            max_error(y_test, y_pred_test),
            mean_absolute_percentage_error(y_test, y_pred_test)
        ]
    }

    # Criando um DataFrame para melhor visualização

    metrics_df = pd.DataFrame(metrics_reg)

    # Formatando os valores para melhor legibilidade

    metrics_df['Treino'] = metrics_df['Treino'].apply(lambda x: f"{x:.4f}")
    metrics_df['Teste'] = metrics_df['Teste'].apply(lambda x: f"{x:.4f}")

    # Exibindo as métricas

    print(metrics_df)

# Exemplo de uso da função

gerar_metricas_regressao(y_test, y_pred_test, y_train, y_pred_train, X_train)
```

Neste bloco, crio uma função chamada `adjusted_r2` para calcular o R² ajustado, que é uma métrica que avalia a qualidade do modelo de regressão, levando em conta o número de variáveis. Para isso, eu uso os valores esperados e previstos, além do conjunto de dados de treino.

Depois, desenvolvi outra função chamada `gerar_metricas_regressao`, que gera várias métricas de desempenho para os dados de treino e teste. Nela, calculo métricas como R² ajustado, erro absoluto médio (MAE), erro quadrático médio (MSE), raiz do erro quadrático médio (RMSE), erro percentual absoluto médio (MAPE) e outras.

As métricas são organizadas em um dicionário e, em seguida, transformadas em um DataFrame do Pandas para facilitar a visualização. Também formatei os valores para que fiquem mais legíveis. Por fim, a função exibe as métricas e as retorna como um DataFrame.

Eu utilizei essa função com os dados de teste e treino para obter as métricas de desempenho do meu modelo.

Ao analisar as métricas de desempenho do meu modelo base line, percebo que ele apresenta sinais de overfitting. Isso acontece quando o modelo se ajusta muito bem aos dados de treinamento, mas não consegue generalizar para novos dados, como os dados de teste.

No meu caso, o modelo teve um desempenho excelente nos dados de treinamento, o que é um bom sinal, mas, ao avaliar os dados de teste, notei uma queda significativa na performance. Isso indica que, embora o modelo tenha aprendido bem as características dos dados de treinamento, ele não conseguiu aplicar esse aprendizado de forma eficaz em dados que não viu antes.

```python
# Definindo a estrutura da Rede Densa

def build_model(hidden=2, n_neurons=4, learning_rate=3e-3):
    model = Sequential()
    model.add(Dense(n_neurons, activation="tanh", kernel_regularizer=L2(), input_shape=X_train_norm.shape[1:]))
    for layer in range(hidden):
        model.add(Dense(n_neurons, activation="tanh", kernel_regularizer=L2()))
    model.add(Dense(1, activation="relu", kernel_regularizer=L2()))
    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(loss="mse", optimizer=optimizer, metrics=['mae', 'mape', 'r2_score'])

    return model

# Envolvendo o modelo com KerasRegressor

keras_reg = KerasRegressor(model=build_model, random_state=42)

# Treinando o modelo

with tf.device('/GPU:0'):
    history = keras_reg.fit(X_train_norm, y_train, epochs=400, batch_size=64,
                      validation_data=(X_test_norm, y_test),
                      callbacks=[keras.callbacks.EarlyStopping(patience=5)])
```

Neste código, inicio definindo a estrutura de uma rede neural densa com a função `build_model`, que permite personalizar a arquitetura do modelo. Começo criando um modelo sequencial com `Sequential()`, adicionando uma camada densa inicial com um número especificado de neurônios, ativação "tanh" e regularização L2. A forma de entrada é baseada nos dados de treinamento normalizados.

Em seguida, uso um loop para adicionar camadas ocultas, todas com a mesma configuração. Na camada de saída, coloco um único neurônio com ativação "relu", adequada para problemas de regressão.

Compilo o modelo com a função de perda "mean squared error" (mse) e o otimizador Adam, além de incluir métricas como MAE, MAPE e R² para avaliar o desempenho.

Depois, envolvo o modelo com `KerasRegressor`, facilitando sua integração com o scikit-learn. Para treinar, utilizo um bloco `with` para garantir que o treinamento ocorra na GPU, chamando o método `fit` com os dados de treinamento e validação, além de definir o número de épocas (400) e o tamanho do lote (64).

Incluo um callback de `EarlyStopping` com paciência de 5, que interrompe o treinamento se não houver melhoria na validação, ajudando a evitar overfitting. Essa abordagem me permite construir e treinar o modelo de forma eficiente, ajustando parâmetros para otimizar o desempenho.

```python
# Plotando histórico de treinamento

def plot_history(history):

    # Acessando o dicionário history diretamente

    history_data = history.history_

    fig, axs = plt.subplots(2, 2, figsize=(20, 10))

    # Plotando a perda (loss)

    axs[0, 0].plot(history_data['loss'], 'r-', label='train loss')
    axs[0, 0].plot(history_data['val_loss'], 'b--', label='test loss')
    axs[0, 0].set_title('Model Loss')
    axs[0, 0].set_xlabel('Epochs')
    axs[0, 0].set_ylabel('Loss')
    axs[0, 0].legend()
    axs[0, 0].grid(False)

    # Plotando o MAE

    axs[0, 1].plot(history_data['mean_absolute_error'], 'r-', label='train mae')
    axs[0, 1].plot(history_data['val_mae'], 'b--', label='test mae')
    axs[0, 1].set_title('Model MAE')
    axs[0, 1].set_xlabel('Epochs')
    axs[0, 1].set_ylabel('MAE')
    axs[0, 1].legend()
    axs[0, 1].grid(False)

    # Plotando o MAPE

    axs[1, 0].plot(history_data['mean_absolute_percentage_error'], 'r-', label='train mape')
    axs[1, 0].plot(history_data['val_mape'], 'b--', label='test mape')
    axs[1, 0].set_title('Model MAPE')
    axs[1, 0].set_xlabel('Epochs')
    axs[1, 0].set_ylabel('MAPE')
    axs[1, 0].legend()
    axs[1, 0].grid(False)

    # Plotando o R²

    axs[1, 1].plot(history_data['r2_score'], 'r-', label='train R²')
    axs[1, 1].plot(history_data['val_r2_score'], 'b--', label='test R²')
    axs[1, 1].set_title('Model R² Score')
    axs[1, 1].set_xlabel('Epochs')
    axs[1, 1].set_ylabel('R² Score')
    axs[1, 1].legend()
    axs[1, 1].grid(False)

    plt.tight_layout()
    plt.show()
```

Seguindo, eu criei uma função chamada `plot_history`, que tem como objetivo visualizar o desempenho do meu modelo Keras durante o treinamento. A função recebe como entrada o objeto `history`, que contém as métricas coletadas ao longo das épocas.

Primeiro, acesso os dados de histórico através de `history.history`, que é um dicionário que armazena as métricas de perda e outras que eu defini, como MAE, MAPE e R². Em seguida, utilizo a biblioteca Matplotlib para criar uma figura com quatro subplots, organizados em duas linhas e duas colunas.

No primeiro subplot, ploto a perda do modelo, tanto para os dados de treinamento quanto para os de validação, usando linhas vermelhas e azuis, respectivamente. Isso me permite observar como a perda evolui ao longo das épocas.

No segundo subplot, faço o mesmo para o erro absoluto médio (MAE), que me ajuda a entender a precisão do modelo. O terceiro subplot exibe o erro percentual absoluto médio (MAPE), que é útil para avaliar a performance em termos percentuais.

Por fim, no quarto subplot, ploto o coeficiente de determinação (R²), que indica a qualidade do ajuste do modelo. Cada gráfico inclui títulos, rótulos nos eixos e legendas para facilitar a interpretação.

Ao final, uso `plt.tight_layout()` para garantir que os gráficos não se sobreponham, melhorando a apresentação visual. Por último, chamo `plt.show()` para exibir os gráficos.
